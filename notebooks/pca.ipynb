```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Chargement
X_original = pd.read_csv('../data/dataset.csv') # Adapte le nom du fichier

# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_original)
# Visualisation (Ajoute une cellule markdown pour ton observation)
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.5)
plt.title("Projection PCA")
plt.show()

# Export
pd.DataFrame(X_pca).to_csv('../outputs/pca_emb_2d.csv', index=False, header=False)
```
