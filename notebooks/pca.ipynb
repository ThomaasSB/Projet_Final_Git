```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

X_original = pd.read_csv('../data/city_lifestyle_dataset.csv')

X_numerique = X_original.select_dtypes(include='number')

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_numerique)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.5, c='blue')
plt.title("Projection PCA des données (2D)")
plt.xlabel("Composante Principale 1")
plt.ylabel("Composante Principale 2")
plt.grid(True)
plt.show()

pd.DataFrame(X_pca).to_csv('../outputs/pca_emb_2d.csv', index=False, header=False)
print("Fichier pca_emb_2d.csv dans output ")
```
