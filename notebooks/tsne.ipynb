```python
plt.figure(figsize=(12, 8)) 

# Création du nuage de points
sns.scatterplot(
    x=X_tsne_2d[:, 0], 
    y=X_tsne_2d[:, 1],
    hue=y, # Ceci va colorer les points en fonction du pays
    palette="tab20", # La palette 'tab20' est excellente quand on a beaucoup de catégories distinctes
    alpha=0.8,
    s=60 # Taille des points
)

plt.title('Projection t-SNE des Villes selon leurs Caractéristiques', fontsize=16)
plt.xlabel('Dimension t-SNE 1')
plt.ylabel('Dimension t-SNE 2')


plt.legend(title='Pays', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
